```python
import os
import csv

# Define the directory where your dataset is stored
dataset_directory = 'DATA/Dataset/iDesigner/designer_image_train_v2_cropped'

# Define the output CSV file name
output_csv_file = 'DATA/Dataset/iDesigner/metadata_cropped.csv'

# Function to traverse the directory and write metadata to CSV
def create_metadata_csv(dataset_dir, output_file):
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['relative_path', 'artist_name'])

        for root, dirs, files in os.walk(dataset_dir):
            for file in files:
                if file.endswith('.jpg') or file.endswith('.png'):  # Adjust file extensions as needed
                    relative_path = os.path.relpath(os.path.join(root, file), dataset_dir)
                    artist = os.path.basename(root)
                    writer.writerow([relative_path, artist])

# Call the function to create metadata CSV
create_metadata_csv(dataset_directory, output_csv_file)

print("Metadata CSV created successfully.")
```

```python
import pandas as pd
import torch
influences = torch.load('DATA/Dataset/iDesigner/dict_influence')
df = pd.read_csv('DATA/Dataset/iDesigner/metadata_cropped.csv')
df['influenced_by'] = df['artist_name'].map(influences)
df.dropna(subset=['influenced_by'], inplace=True)
all_artist_names = set(df['artist_name'])
df['influenced_by'] = df['influenced_by'].apply(lambda artists_list: [artist for artist in artists_list if artist in all_artist_names])
df = df[df['influenced_by'].apply(len)>0].reset_index(drop=True)
df.dropna(subset=['influenced_by'], inplace=True)
df.to_pickle('DATA/Dataset/iDesigner/idesigner_influences_cropped.pkl')
```
